# Competencia Doble Rendija

En la física moderna, el experimento de la doble rendija es una demostración de que la luz y la materia pueden mostrar características tanto de ondas como de partículas definidas clásicamente; además, muestra la naturaleza fundamentalmente probabilística de los fenómenos de la mecánica cuántica.



### Marco teórico previo

Primero para comprender los que sucede en el experimento hay que comprender
de que trata este mismo este fue realizado por primera vez por Thomas Young en
1801y demostró que la luz exhibe un comportamiento de onda a través de la
interferencia, lo que dio paso al concepto de onda-partícula concepto fundamenta
de la física moderna.



### Procedimiento

Para proceder con el desarrollo del el experimento de la doble rendija procedimos
con los siguientes materiales: -Una caja de cartón, papel Aluminio y un láser.
Al proceder de esta actividad la intención es poder reproducir el patrón de
interferencia que se debería evidenciar en la imagen que resulta del láser después
de pasar por la doble rendija.

Con estos materiales los que vamos a hacer unos cortes muy delgados y con una distacia entre si muy corta, ya teniendo esta lamina la pegaremos en hueco previamente hecho en la caja y al lado opuesto de esta lamina haremos otro hueco a la caja para poder soportar un laser en este lugar apuntando a los cortes de la lamina de aluminio.



Ahora estando en un lugar oscuro podemos prender el laser y ver como se forma un patron de lineas (En el centro se ve una luz mas fuerte por la potencia del laser).




### Explicación de los resultados

En el experimento primeramente preparamos todo, las rendijas de aluminio y los
agujeros en la caja para al momento de probar el experimento pudimos notar que la
luz al atravesar las rendijas efectivamente presenta un patrón parecido al que
muestra en el video y que podemos que explicar que sucede por que la luz que pasa
por cada rendija actúa como una fuente de ondas que se superponen generando
regiones donde hay interferencias diferentes.
Interferencia constructiva: Donde se producen franjas brillantes al estar las ondas
en fase.
Interferencia destructiva: donde se producen franjas oscuras, debido a que las
ondas están en oposición de fase.
En una explicación cuántica, cada partícula es descrita por una función de onda,
que representa un estado de superposición de todas las posibles trayectorias, antes
de ser medida su estado es una combinación de ambos caminos.
El entrelazamiento y la información cuánticos también son elementos importantes,
donde dos partículas o qubits comparten estado cuántico, de modo que la medición
de una afecta el estado de la otra, en la dobles rendija se ha explorado un fenómeno
similar mediante versiones cuánticas de experimento donde los fotones
entrelazados pueden mostrar o no interferencia de una afecta instaniamente el
estado de la otra, sin importar distancia, Esto sugiere que la información juega un
papel crucial en la mecánica cuántica, tanto en el colapso de la función de onda
como en la computación cuántica, donde los qubits entrelazados pueden realizar
cálculos imposibles para computadoras clásicas.
